```python
from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
name = pd.read_csv("HULTPRIZE.csv")
name.head()
lis = name['name'].to_list()
print(lis)
```

```python
image = Image.open('competitors.png')
```

```python
plt.imshow(image)
```

```python
def certif(name):
    copy = image.copy()
    draw = ImageDraw.Draw(copy)
    font = ImageFont.truetype("Century Bold Italic.ttf", 150)

    draw.text((1200,1200),name,(0,0,0),font=font)
    nom = 'certif/' + name + '.png'
    copy.save(nom)
```

```python
for i in lis:
    certif(i)
```
